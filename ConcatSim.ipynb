```python
directory = os.path.abspath("/content/drive/MyDrive/HS651_CAMs_systematic_review_automation/QA(notC)Similarity/")
pickleFiles = glob.glob(os.path.join(directory, '*cleanQA.pkl'))
```

```python
#for PMCpickle in pickleFiles:
    #print(str(PMCpickle[82:92]))
```

```python
for PMCpickle in pickleFiles:
  picklef = pd.read_pickle (PMCpickle)
  PMCr = str(PMCpickle[82:92])
  txgroup = picklef.loc[picklef['Question'] == 'what are the treatment groups?', 'SimilarityQA'].mean()
  sleepm = picklef.loc[picklef['Question'] == 'what is the scale of sleep measure?', 'SimilarityQA'].mean()
  scaledep = picklef.loc[picklef['Question'] == 'what is the scale of depression measure?', 'SimilarityQA'].mean()
  scaleanx = picklef.loc[picklef['Question'] == 'what is the scale of anxiety measure?', 'SimilarityQA'].mean()
  longint = picklef.loc[picklef['Question'] == 'how long is the intervention?', 'SimilarityQA'].mean()
  df.loc[df.PMCID == PMCr,'TreatmentGroupSimQA'] = txgroup
  df.loc[df.PMCID == PMCr,'SleepMeasureSimQA'] = sleepm
  df.loc[df.PMCID == PMCr,'ScaleDepressionSimQA'] = scaledep
  df.loc[df.PMCID == PMCr,'ScaleAnxietySimQA'] = scaleanx
  df.loc[df.PMCID == PMCr,'LengthInterventionSimQA'] = longint
```

```python
df = df.drop(['TreatmentGroupSim','SleepMeasureSim','ScaleDepressionSim','ScaleAnxietySim','LengthInterventionSim'],axis = 1)
```
