```python
import re
import pickle
import spacy
import numpy as np
from spacy.lang.en import English
```

```python
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
```

```python
nlp = spacy.load('en')
```

```python
nlp.Defaults.stop_words -= {"will","had","was","were","did"}
```

```python
new_stopwords = spacy_stopwords
```

```python
def clean_text(x):
  x = x.lower().split()
  new_list = []
  for word in x:
    if word not in new_stopwords: 
      new_list.append(word)
  return " ".join(new_list)
```

```python
# to get only text in between introduction and references
# find (first occurence) , rfind (find last  occurence) are python string methods
def preprocess(x):
    start = "introduction"
    end = "reference"
    x =  x[x.find(start)+len(start):x.rfind(end)]
    return x
```

```python
for one_txt in TxtFiles:
    ft = open(one_txt,'rt')
    txt = ft.read()
    txt = clean_text(txt)
    txt = preprocess(txt)
    savepath = "/content/drive/MyDrive/fulltexts/"
    filenameString = str(one_txt[:-8]) + "clean"+".txt"
    completeName = os.path.join(savepath,filenameString)
    text_file = open(completeName, "w")
    text_file.write(txt)
    ft.close()
```

A commonly used strategy to split long documents, especially in the field of Question Answering, is the sliding window approach. If split_length=10 and split_overlap=3, your documents will look like this:

doc1 = words[0:10]
doc2 = words[7:17]
doc3 = words[14:24]

```python
from haystack.file_converter.txt import TextConverter
#from haystack.file_converter.pdf import PDFToTextConverter
from haystack.preprocessor.preprocessor import PreProcessor
from haystack.retriever.sparse import ElasticsearchRetriever
#from haystack.pipeline import ExtractiveQAPipeline
converter = TextConverter(remove_numeric_tables=True, valid_languages=["en"])


savepath = "/content/drive/MyDrive/Abstxt20_158/"

for abs in abstxt:
  if os.path.getsize(abs) > 0:
     document_store.delete_all_documents()
     doc_txt = converter.convert(file_path=abs,meta=None)  
     preprocessor_sliding_window = PreProcessor(
       split_overlap=1,
       split_length=500,
       split_respect_sentence_boundary=True,
       clean_empty_lines=True,
       clean_whitespace=True,
       clean_header_footer=True)


     docs_sliding_window = preprocessor_sliding_window.process(doc_txt)
     document_store.write_documents(docs_sliding_window)
     retriever = ElasticsearchRetriever(document_store=document_store)
     from haystack.pipeline import ExtractiveQAPipeline
     pipe = ExtractiveQAPipeline(reader, retriever)

     prediction1 = pipe.run(query="what is the intervention/treatment group?", top_k_retriever=10, top_k_reader=5)
     predictiondata1 = [[prediction1['query'],prediction1['answers'][1]['answer'],prediction1['answers'][1]['context']],
        [prediction1['query'],prediction1['answers'][2]['answer'],prediction1['answers'][2]['context']],
        [prediction1['query'],prediction1['answers'][3]['answer'],prediction1['answers'][3]['context']]]
     prediction2 = pipe.run(query="what is the scale of sleep measure?", top_k_retriever=10, top_k_reader=5)
     predictiondata2 = [[prediction2['query'],prediction2['answers'][0]['answer'],prediction2['answers'][0]['context']],
        [prediction2['query'],prediction2['answers'][1]['answer'],prediction2['answers'][1]['context']],
        [prediction2['query'],prediction2['answers'][2]['answer'],prediction2['answers'][2]['context']]]
     prediction3 = pipe.run(query="what is the scale of depression measure?", top_k_retriever=10, top_k_reader=5)
     predictiondata3 = [[prediction3['query'],prediction3['answers'][1]['answer'],prediction3['answers'][1]['context']],
        [prediction3['query'],prediction3['answers'][2]['answer'],prediction3['answers'][2]['context']],
        [prediction3['query'],prediction3['answers'][3]['answer'],prediction3['answers'][3]['context']]]
     prediction4 = pipe.run(query="what is the scale of anxiety measure?", top_k_retriever=10, top_k_reader=5)
     predictiondata4 = [[prediction4['query'],prediction4['answers'][1]['answer'],prediction4['answers'][1]['context']],
        [prediction4['query'],prediction4['answers'][2]['answer'],prediction4['answers'][2]['context']],
        [prediction4['query'],prediction4['answers'][3]['answer'],prediction4['answers'][3]['context']]]
     prediction5 = pipe.run(query="how long is the intervention?", top_k_retriever=10, top_k_reader=5)
     predictiondata5 = [[prediction5['query'],prediction5['answers'][1]['answer'],prediction5['answers'][1]['context']],
        [prediction5['query'],prediction5['answers'][2]['answer'],prediction5['answers'][2]['context']],
        [prediction5['query'],prediction5['answers'][3]['answer'],prediction5['answers'][3]['context']]]
     df1 = pd.DataFrame(predictiondata1, columns = ['Question', 'Answer','Context'])
     df2 = pd.DataFrame(predictiondata2, columns = ['Question', 'Answer','Context'])
     df3 = pd.DataFrame(predictiondata3, columns = ['Question', 'Answer','Context'])
     df4 = pd.DataFrame(predictiondata4, columns = ['Question', 'Answer','Context'])
     df5 = pd.DataFrame(predictiondata5, columns = ['Question', 'Answer','Context'])
     dflist = [df1,df2,df3,df4,df5]
     queriesdfs = pd.concat(dflist)
     filenameString = str(abs[:-6]) +"Abs"+".pkl"
     completeName = os.path.join(savepath,filenameString)
     queriesdfs.to_pickle(completeName)
```
