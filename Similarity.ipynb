```python
import pandas as pd
import scispacy
import spacy
nlp = spacy.load("en_core_sci_lg")
```

```python
def get_similarity(term1, term2):
    token1 = nlp(term1)
    token2 = nlp(term2)
    return token1.similarity(token2)
```

```python
import os
import glob
```

```python
directory = os.path.abspath("/content/drive/MyDrive/Acceptable_clean_text_as1")
pickleFiles = glob.glob(os.path.join(directory, '*clean.pkl'))
```

```python
for PMCpickle in pickleFiles:
    print(str(str(PMCpickle[50:60])))
```

```python
savepath = "/content/drive/MyDrive/Acceptable_clean_text_as1"
for PMCpickle in pickleFiles:
  picklef = pd.read_pickle (PMCpickle)
  PMCr = str(PMCpickle[50:60])
  picklef = picklef.reset_index(drop=True)
  #picklef = picklef.dropna() # don't drop na for similarities between Question and Answer
  picklef['SimilarityQA'] = picklef.apply(lambda x: get_similarity(x['Question'], x['Answer']), axis = 1)
#print(picklef)
  stringpkl = str(PMCr) + 'cleanQA.pkl'
  completeName = os.path.join(savepath, stringpkl)
  picklef.to_pickle(completeName)
```
